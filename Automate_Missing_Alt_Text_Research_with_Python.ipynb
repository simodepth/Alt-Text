##Automate Missing Alt Text Research with Python


---

Alt Text is a crucial requirement when it comes to Image optimisation in SEO. Although the venture of MultiSearch in Google may soon prompt users formulate queries from an expanded range of media, Alt Text still represents a milestone in images optimisation as it provides search engines and the public with semantic info scoped to improve accessibility and searchability.

Based on the URLs list you decide to investigate, this Python script is coded to deliver a CSV file in charge of proving the existence of an Alt Text on every single images you submitted.

#Recommended Steps
1. Crawl a website with a tool of your choice and export URLs in xlsx
2. Upload a file containing a list of URLs you want to investigate
3. Make sure you list URLs within the first column and name it "URL".






```python
#@title Import Packages
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#@title Upload List of URLs in an XLSX file 
df = pd.read_excel("/content/Untitled spreadsheet.xlsx", sheet_name=0) #@param {type:"string"}
```

```python
#@title Reading URLs and parsing HTML
df["Immagini"] = ""

for index, row in df.iterrows():
    try:
      
        pagina = requests.get(row[0])
        contenutoPagina = pagina.content
        contenutoParsato = BeautifulSoup(contenutoPagina,"html.parser")
        
        try:
            immagini = contenutoParsato.find_all("img")
            ls_img = []
            for immagine in immagini:
                try:
                    url = immagine.get("src")
                    if url != "" and url != None:
                        try:
                            alt = immagine.get("alt")
                            if alt == None or alt == "":
                                ls_img.append(url)
                        except:
                            ls_img = ls_img
                except:
                    ls_img = ls_img
            if len(ls_img) > 0:
                row["Images"] = ls_img
            else:
                row["Images"] = "There are no images without Alt tag"
        except:
            row["Images"] = "Error"
        
    except:
        row["Images"] = "General Error"
```

```python
#@title Amend file name output 
try:
    df.to_excel("Missing-Alt-Text.xlsx", index=False)
    print("File saved")
except:
    print("Error save")
```
